```python
import pyscf
from pyscf.tools import cubegen
import matplotlib.pyplot as plt
```

```python
# molpath = "../hamiltonians/N2/N2_bond1.1000.chk"
# mol = pyscf.lib.chkfile.load_mol(molpath)

mol = pyscf.gto.Mole()
mol.atom = "N 0 0 0; N 2.0 0 0"
mol.basis = "sto-3g"
mol.symmetry = "D2h"
mol.build()

mf = pyscf.scf.RHF(mol)

# mf.chkfile = "../hamiltonians/N2/N2_2.0000_D2h.chk"
mf.kernel()
```

```python
mf.mo_energy
```

```python
mol2 = pyscf.gto.Mole()
mol2.atom = "N 0 0 0"
mol2.basis = "sto-3g"
mol2.spin = 3
mol2.build()

mf2 = pyscf.scf.RHF(mol2)

mf2.kernel()

mf2.mo_energy
```

```python
mol.groupname
```

```python

orb_syms = pyscf.symm.label_orb_symm(mol, mol.irrep_name, mol.symm_orb, mf.mo_coeff)
print(orb_syms)

# ['Ag' 'B1u' 'Ag' 'B1u' 'B2u' 'B3u' 'Ag' 'B2g' 'B3g' 'B1u']
```

```python
plt.imshow(mf.mo_coeff, cmap="PuOr", vmin=-1, vmax=1)
plt.yticks(range(mol.nao), mol.ao_labels())
plt.xticks(range(mol.nao), range(mol.nao))
plt.colorbar()
plt.show()
```
